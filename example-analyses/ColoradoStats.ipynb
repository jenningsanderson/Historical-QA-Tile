```python
import json, matplotlib
import pandas as pd; import numpy as np
%matplotlib inline
```

```python
df = pd.read_json('minute_counts_co.json','r')
df.columns = ['minutes', 'count']
```

```python
df['minutes'] = pd.to_numeric(df['minutes'],errors='coerce')
```

```python
df['hours'] = df.minutes.apply(lambda x: x/60)
df['days']  = df.hours.apply(lambda x: x/24)
```

```python
df.hours.plot(kind='hist')
```

```python
df.days.plot(kind='hist')
```

```python
ax = df.query('days < 365').days.plot(kind="hist", bins=365, figsize=(15,10))
ax.set_xlim(0,365)
```

```python
df.query('hours < 24').hours.plot(kind="hist", bins=24)
```

```python
df.query('minutes < 60').minutes.plot(kind="hist",bins=60)
```
